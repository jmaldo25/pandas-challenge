```python
# Setting calculations and buckets for unique items, average price, number of purchases, and total revenue 
unique_items = len(pd.unique(pdata2['Item ID']))
avg_price = pdata2["Price"].mean()
total_purchases = len(pd.unique(pdata2['Purchase ID']))
total_rev = pdata2["Price"].sum()
round_price = round(avg_price, 2)

# Need to convert data into df
purchase_analysis_df = {'Number of Unique Items' : [unique_items],
                       'Average Price' : [round_price],
                       'Number of Purchases' : [total_purchases],
                       'Total Revenue' : [total_rev]
                       }
pa_df = pd.DataFrame(purchase_analysis_df)
pa_df.style.format({'Average Price': '${0:,.2f}', 'Total Revenue': '${0:,.2f}'})
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Setting up groupby
spenders = pdata2.groupby("SN")

# Setting up calculations
spender_count = spenders["Purchase ID"].count()

avg_price_spend = spenders["Price"].mean()

total_prch_spend = spenders["Price"].sum()

top_spenders = pd.DataFrame({"Purchase Count": spender_count,
                            "Average Purchase Price": avg_price_spend,
                            "Total Purchase Value": total_prch_spend})
top_spenders_ascend = top_spenders.sort_values(["Total Purchase Value"], ascending=False).head()

top_spenders_df = top_spenders_ascend.style.format({'Average Purchase Price': '${0:,.2f}', 
                                                    'Total Purchase Value': '${0:,.2f}',})
top_spenders_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Create data frame for popular items
items_df = pdata2[["Item ID", "Item Name", "Price"]]

# Need to groupby Item ID and Item Name
items_grp = items_df.groupby(["Item ID", "Item Name"])

# Need calculations for Purchase Count, Item Price, Total Purchase Value
prch_count = items_grp["Item ID"].count()

item_price = items_grp["Price"].mean()

tpv = items_grp["Price"].sum()

# Create dataframe for calculations
total_items_df = pd.DataFrame({"Purchase Count": prch_count,
                              "Item Price": item_price,
                              "Total Purchase Value": tpv})
# Clean dataframe for descending order
items_df_desc = total_items_df.sort_values(["Purchase Count"], ascending=False).head()

items_df_clean = items_df_desc.style.format({'Item Price': '${0:,.2f}', 
                                            'Total Purchase Value': '${0:,.2f}',})
items_df_clean
```
